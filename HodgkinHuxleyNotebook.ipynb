```python
from HodgkinHuxleyEquations import * 
from HodgkinHuxleyPlot import *
```

## Physiological state 

```python
# Simulation parameters
dt = 0.001 #Time-step (ms)
tmax = 200 #Duration of simulation (ms)
n_steps = int(tmax / dt) #Number of steps
sim_time = np.linspace(0, tmax, n_steps)
```

### Constant $I_{input}$

```python
# Input current (stimulation) (ÂµA/cm**2)
I_cst = np.zeros(n_steps) 
I_cst[:n_steps] = 12
```

```python
# Simulation
results_physio = runSimulation(I_cst, sim_time)
```

```python
# Plot Voltage Over Time
plotVoltage(results_physio["V"], sim_time)
```

```python
# Zoom on 1 action potential 
voltage_AP = results_physio["V"][20000:45000]
sim_time_AP = sim_time[20000:44999]

# Plot
plotVoltage(V=voltage_AP, sim_time=sim_time_AP)
```

### Variable $I_{input}$

```python
# Input current (stimulation) (ÂµA/cm**2) 
I_var = np.zeros(n_steps) 
I_var[10000:20000] = 12
I_var[40000:70000] = 12
I_var[100000:130000] = 30
I_var[160000:190000] = 60
```

```python
# Simulation
results_physio = runSimulation(I_var, sim_time)
```

```python
# Plot
plotAll(results_physio, I_var, sim_time)
```
